# Whirpool Effect

swimmers stated that the pool used for the 2013 fina world swimming chamionship had a whirpool effect, lets investgate the data to verify this

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from pandasgui import show
```

```python
finals = pd.read_csv('medalsbylane.csv')
finals
```

```python
finals['Rank'] = finals['Rank'].astype('int64')
finals.info()
```

first we check if higher lanes garnered more medals than lower lanes for this edition. keep in mind we will check only competitions that are fifty meters in length

```python
finals['lane type'] = np.where(finals['Lane'] <= 3, 'lanes 1-3', np.where(finals['Lane'] >= 6, 'lanes 6-8', 'middle lanes'))
```

```python
medals = finals[(finals['Rank']<=3) & (finals['lane type'] != 'middle lanes')].groupby(['edition','lane type'])['Lane'].count()
medals_pivot = medals.unstack()
m1 = pd.DataFrame(medals_pivot.to_records())
m1
```

```python
edition = m1.edition
low_lanes = m1['lanes 1-3']
high_lanes = m1['lanes 6-8']

X_axis = np.arange(len(edition))

plt.bar(X_axis - 0.2, low_lanes, 0.4, label = 'low lanes')
plt.bar(X_axis + 0.2, high_lanes, 0.4, label = 'high lanes')

plt.xticks(X_axis, edition)
plt.xlabel("edition")
plt.ylabel("Number of Medals")
plt.title("Number of Medals by lane type")
plt.legend()
plt.show()
```

2013 clearly has a skew. now lets dive deeper into the date from 2013 itself

```python
fina_thirteen = pd.read_csv('2013_FINA.csv')
fina_thirteen
```

```python
fina_thirteen.info()
```

lets compare the time for swimmers who swam in a high/low lane in the semi and the other lane type for the final

```python
fina_thirteen['lane type'] = np.where(fina_thirteen['lane'] <= 3, 'low lane', np.where(fina_thirteen['lane'] >= 5, 'high lane', 'middle lanes'))
```

```python
fina_thirteen['eid'] = fina_thirteen['eventid'].astype('str').str[-2:].astype('int64')
fina_thirteen_filtered = fina_thirteen[(fina_thirteen['distance'] == 50) & ((fina_thirteen['round'] == 'SEM') | (fina_thirteen['round'] == 'FIN')) & (fina_thirteen['lane type'] != 'middle lanes')]
fina_thirteen_filtered_sems = fina_thirteen_filtered[fina_thirteen_filtered['round'] == 'SEM']
fina_thirteen_filtered_fin = fina_thirteen_filtered[fina_thirteen_filtered['round'] == 'FIN']
joined = pd.merge(fina_thirteen_filtered_sems, fina_thirteen_filtered_fin, on=['eid', 'athleteid'])
candidates = joined[joined['lane type_x'] != joined['lane type_y']]
```

```python
candidates['numerator'] = np.where(candidates['lane type_x'] == 'low lane', candidates['swimtime_x'] - candidates['swimtime_y'], candidates['swimtime_y'] - candidates['swimtime_x'])
candidates['denominator'] = np.where(candidates['lane type_x'] == 'low lane', candidates['swimtime_x'], candidates['swimtime_y'])
candidates['f'] = candidates['numerator'] / candidates['denominator']
show(candidates)
```

```python
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""
    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n

    return x, y
```

```python
x,y = ecdf(candidates['f'])
plt.plot(x,y,marker='.', linestyle='none')
plt.xlabel('f')
plt.ylabel('ecdf')
plt.show()
```

```python
def draw_bs_reps(data, func, size=1):
    """Draw bootstrap replicates."""

    # Initialize array of replicates: bs_replicates
    bs_replicates = np.empty(size)

    # Generate replicates
    for i in range(size):
        bs_replicates[i] = bootstrap_replicate_1d(data, func)

    return bs_replicates
def bootstrap_replicate_1d(data, func):
    """Generate bootstrap replicate of 1D data."""
    bs_sample = np.random.choice(data, len(data))
    return func(bs_sample)
```

```python
# Compute the mean difference: f_mean
f_mean = np.mean(candidates['f'])

# Draw 10,000 bootstrap replicates: bs_reps
bs_reps = draw_bs_reps(candidates['f'], np.mean, 10000)

# Compute 95% confidence interval: conf_int
conf_int = np.percentile(bs_reps, [2.5, 97.5])

# Print the result
print("""
mean frac. diff.: {0:.5f}
95% conf int of mean frac. diff.: [{1:.5f}, {2:.5f}]""".format(f_mean, *conf_int))
```

```python
f_shift = candidates['f'] - f_mean

# Draw 100,000 bootstrap replicates of the mean: bs_reps
bs_reps = draw_bs_reps(f_shift,np.mean, 100000)

# Compute and report the p-value
p_val = np.sum(bs_reps >= f_mean) / 100000
print('p =', p_val)
```
